```python
import numpy as np
import cv2
import numpy as np
import time
import pyautogui
import keras
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.models import load_model
import matplotlib.patches as patches
import keras.utils as image
import pyautogui
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
```

```python
model = load_model("rps.h5")
```

```python
cap = cv2.VideoCapture(0)

class_labels = ["Paper", "Rock", "Scissor"]

cooldown_seconds = 5
last_slide_change_time = 0

while True:
    ret, frame = cap.read()

    if not ret:
        break

    try:
        cv2.rectangle(frame, (50, 50), (400, 400), (255, 255, 255), 2)

        roi = frame[50:400, 50:400]
        roi = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
        roi = cv2.resize(roi, (150, 150))

        x = image.img_to_array(roi)
        x = np.expand_dims(x, axis=0)
        images = np.vstack([x])

        pred = model.predict(images)
        predicted_class_index = np.argmax(pred[0])
        predicted_class_label = class_labels[predicted_class_index]

        print(predicted_class_label)

        if predicted_class_label == "Paper":
            cv2.putText(frame, "Predicted: " + str(predicted_class_label),
                    (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255, 255, 255), 2, cv2.LINE_AA)
        elif predicted_class_label == "Scissor":
            cv2.putText(frame, "Predicted: " + str(predicted_class_label),
                    (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255, 255, 255), 2, cv2.LINE_AA)
        
        
        current_time = time.time()
        if current_time - last_slide_change_time > cooldown_seconds:
            if predicted_class_label == "Paper":
                pyautogui.press("right")
                last_slide_change_time = current_time
            elif predicted_class_label == "Scissor":
                pyautogui.press("left")
                last_slide_change_time = current_time
        cv2.imshow("Rock Paper Scissors", frame)

        if cv2.waitKey(10) == ord('q'):
            break

    except Exception as e:
        print(f"Error processing the frame: {str(e)}")

cap.release()
cv2.destroyAllWindows()
```
